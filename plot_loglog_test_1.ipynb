## Tutorial: plotting functions in python
Example of plotting multiple data sets in 1 figure using matplotlib

```python
#import
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
#create test functions to plot [f(n), g(n) functions of time variable n]

N=100 #maximum time to consider

#time array (n=0, 1, 2, ...)
n_vals=np.linspace(0, N, N+1)

#print("n_vals =",n_vals)


#test function #1 f(n)=1/(1+n)
f_vals=1.0/(1.0+n_vals)

#print("f_vals =",f_vals)


#test function #2 g(n)=1/(1+n^(1/2))
g_vals=1.0/(1.0+n_vals**0.5)

#print("g_vals =",g_vals)
```

```python
#plot test functions (data points / scatter plot)

xpoints=n_vals
ypoints=f_vals

marker_size=10 #marker size

plt.scatter(xpoints, ypoints, label = "f(n)", s=marker_size)

xpoints=n_vals
ypoints=g_vals

plt.scatter(xpoints, ypoints, label = "g(n)", marker="x", s=marker_size) #note: can make g(n) points different marker type since f(n)=g(n) when n=0,1

plt.xlabel("Time (n)")
plt.ylabel("Test function: f(n), g(n)")
plt.legend()
plt.show()
```

```python
#plot test functions (log-log plot)

xpoints=n_vals
ypoints=f_vals

plt.loglog(xpoints, ypoints, marker='o', linestyle='None', markersize=5, color='blue', label = "f(n)")

xpoints=n_vals
ypoints=g_vals

plt.loglog(xpoints, ypoints, marker='x', linestyle='None', markersize=5, color='orange', label = "g(n)") #note: can make g(n) points different marker type since f(n)=g(n) when n=0,1

plt.xlabel("Time (n)")
plt.ylabel("Test function: f(n), g(n)")
plt.legend()
plt.show()
```
